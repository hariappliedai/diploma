# Median

- We have seen that mean is measure of central tendecy, but it is easily corrupted by outliers.
- Median is also a mesure of cetral tendency, but is not much effected by outiers.

```
# importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load Iris.csv into a pandas dataFrame.
iris = pd.read_csv("iris.csv")
```

```
iris_setosa = iris[iris['species'] == 'setosa']
iris_virginica = iris[iris['species'] == 'virginica']
iris_versicolor = iris[iris['species'] == 'versicolor']
```

```
print('=======================Median=========================')
print('setosa ----> {}'.format(np.median(iris_setosa['petal_length'])))
print('versicolor ----> {}'.format(np.median(iris_versicolor['petal_length'])))
print('virginica ----> {}'.format(np.median(iris_virginica['petal_length'])))
```
