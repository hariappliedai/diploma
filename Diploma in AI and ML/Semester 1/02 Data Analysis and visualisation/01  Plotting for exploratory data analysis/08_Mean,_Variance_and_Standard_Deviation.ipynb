```
# importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load Iris.csv into a pandas dataFrame.
iris = pd.read_csv("iris.csv")
```

```
iris_setosa = iris[iris['species'] == 'setosa']
iris_virginica = iris[iris['species'] == 'virginica']
iris_versicolor = iris[iris['species'] == 'versicolor']
```

```
print('=======================Mean=========================')
print('setosa ----> {}'.format(np.mean(iris_setosa['petal_length'])))
print('versicolor ----> {}'.format(np.mean(iris_versicolor['petal_length'])))
print('virginica ----> {}'.format(np.mean(iris_virginica['petal_length'])))
print('======================Std Dev ========================')
print('setosa ----> {}'.format(np.std(iris_setosa['petal_length'])))
print('versicolor ----> {}'.format(np.std(iris_versicolor['petal_length'])))
print('virginica ----> {}'.format(np.std(iris_virginica['petal_length'])))
```

One problem with mean is that means gets corrupted due to outliers and th same is applicable for variance and std-dev.

```
print('Actual mean ----> {}'.format(np.mean(iris_setosa['petal_length'])))
print('Corrupted mean ----> {}'.format(np.mean(np.append(iris_setosa['petal_length'],50))))
```
