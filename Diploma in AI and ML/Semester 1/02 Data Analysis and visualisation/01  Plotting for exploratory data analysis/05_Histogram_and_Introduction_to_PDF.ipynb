```
# importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# loading data
iris = pd.read_csv("iris.csv")
```

Up to now we have seen scatter plots in 2D and 3D but what about 1D.

let't draw a scatter plot by taking only a single feature and on the other axis lets consider all zeros.

```
iris_setosa = iris[iris['species'] == 'setosa']
iris_virginica = iris[iris['species'] == 'virginica']
iris_versicolor = iris[iris['species'] == 'versicolor']

plt.scatter(iris_setosa['petal_length'], np.zeros_like(iris_setosa['petal_length']))
plt.scatter(iris_versicolor['petal_length'], np.zeros_like(iris_versicolor['petal_length']))
plt.scatter(iris_virginica['petal_length'], np.zeros_like(iris_virginica['petal_length']))
plt.xlabel('petal_length')
plt.legend(['setosa', 'versicolor', 'virginica'])

plt.show()
```

In the above plot we can see that many points overlap and it is hard to read. So, how about stacking all the points in a range and this idea is called as Histogram

## Histogram

```
sns.FacetGrid(iris, hue='species', height=7) \
    .map(sns.distplot, 'petal_length') \
    .add_legend()
plt.show()
```

- The bocks which are giving the iformation about the density of flowers at a given range of petal_length is histogram.
- The smooth suface trying to approximate the histogram is PDF (probability density function), we will look into this later
