```
# importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load Iris.csv into a pandas dataFrame.
iris = pd.read_csv("iris.csv")
```

## CDF of petal_length of Setosa species

```
iris_setosa = iris[iris['species'] == 'setosa']

counts, bin_edges = np.histogram(iris_setosa['petal_length'],
                                 bins = 10, density = True)
#compute PDF
pdf = counts/(sum(counts))

#compute CDF
cdf = np.cumsum(pdf)  # to compute cummulative sum

print('bin_edges ----> {}'.format(bin_edges))
print('PDF ----> {}'.format(pdf))
print('CDF ----> {}'.format(cdf))
```

```
plt.plot(bin_edges[1:],pdf, )
plt.plot(bin_edges[1:],cdf)
plt.legend(['PDF', 'CDF'])
plt.title('PDF and CDF of petal_length(Setosa)')

plt.show()
```
