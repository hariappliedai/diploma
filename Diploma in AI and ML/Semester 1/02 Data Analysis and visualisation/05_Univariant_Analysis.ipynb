# Univariant Analysis

Univariant analysis means analysing only a single feature. In this notebook lets do a univariant analysis on all the 4 features individually

```
!wget https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv # downloading csv file
```

```
# importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# loading data
iris = pd.read_csv("iris.csv")
```

```
sns.FacetGrid(iris, hue='species', height=5) \
    .map(sns.distplot, 'sepal_length') \
    .add_legend()
plt.show()
```

```

sns.FacetGrid(iris, hue='species', height=5) \
    .map(sns.distplot, 'sepal_width') \
    .add_legend()
plt.show()
```

```

sns.FacetGrid(iris, hue='species', height=5) \
    .map(sns.distplot, 'petal_length') \
    .add_legend()
plt.show()
```

```

sns.FacetGrid(iris, hue='species', height=5) \
    .map(sns.distplot, 'petal_width') \
    .add_legend()
plt.show()
```

<u>Univariant Analysis:</u>
- sepal_length and sepal_width are not of use for classification due to the overlap.
- petal_length and petal_width are useful in classification and if we have to use only a single feature from both of this that would be petal_length because of the well seperated PDF's
